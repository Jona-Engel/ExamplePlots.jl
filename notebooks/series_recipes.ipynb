```julia
using Plots
pyplot(size=(400,250))

# This is a series recipe.
# Create your own series type to act on any valid inputs
@recipe function f(::Type{Val{:my_histogram}}, x, y, z)
    edges, counts = hist(y, d[:bins])
    d[:seriestype] = :bar
    d[:fillalpha] = 0.3
    d[:bar_width] = 0.8mean(diff(edges))
    d[:x] = 0.5(edges[1:end-1] + edges[2:end])
    d[:y] = counts
    ()
end

# a line plot of the data
plot(randn(1000,2), α=0.5)
```

```julia
# our custom histogram!
plot(randn(1000), t=:my_histogram)
```

```julia
# overlapping histograms, with no extra work!
plot(randn(1000,3), t=:my_histogram)
```

```julia
# random colors?
Base.rand(::Type{RGB}, n) = [RGB(rand(),rand(),rand()) for i=1:n]

# why would anyone do this?
@recipe function f(::Type{Val{:funky_bar}}, x, y, z)
    d[:seriestype] = :bar
    d[:fillcolor] = rand(RGB, length(x))
    ()
end

# funkadelic!
plot(rand(10), t=:funky_bar)
```

```julia
# lets redefine a histogram, and give our user the option of funk
@recipe function f(::Type{Val{:funky_histogram}}, x, y, z)
    edges, counts = hist(y, d[:bins])
    d[:seriestype] = (pop!(d, :funky, false) ? :funky_bar : :bar)
    d[:fillalpha] = 0.3
    d[:bar_width] = 0.8mean(diff(edges))
    d[:x] = 0.5(edges[1:end-1] + edges[2:end])
    d[:y] = counts
    ()
end

# booooooorrrriiiiiiing
plot(randn(1000), t=:funky_histogram)
```

```julia
# that's more like it!
plot(randn(1000), t=:funky_histogram, funky=true)
```
