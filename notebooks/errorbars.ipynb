```julia
using Plots
gadfly(leg=false, size=(500,300))
srand(1)

sds = [1, 1/2, 1/4, 1/8, 1/16, 1/32]
n = 10
y = [mean(sd*randn(n)) for sd in sds]
errs = 1.96 * sds / sqrt(n)

plot(y, err = errs)
```

```julia
plot(y, 1:length(y), err = errs,
     orientation = :h,
     marker = stroke(3,:orange))
```
